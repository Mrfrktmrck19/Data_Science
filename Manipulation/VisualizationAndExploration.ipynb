## Doğrusal İlişkinin Gösterilmesi

Scatter plot ile iki sayısal değişkenin arasındaki ilişkiyi "Saçılımsal" olarak gözlemlemiştik. Şimdi bir doğru ile bu ilişkiyi göeceğiz.

```python
# lmplot = lineer model plot
sns.lmplot(x = "total_bill", y = "tip", data = df)
```

Doğrunun yakınlarındaki değişkenler ilişkisi yüksek iken, doğrudan uzak olan değişkenler outlier'dır. Yani sapması yüksek olan, veri setimizde ki aykırı durumlardır.

```python
sns.lmplot(x = "total_bill", y = "tip", hue = "smoker", data = df)
```

Ödenilen total hesap baz alındığında: sigara içen kişilerin verdiği bahşiş, içmeyen kişilerin verdiği bahşişten çok daha fazla. 

Bu tarz yorumu Veri Bilimi için istatistik konusunda ele alacağız.

```python
sns.lmplot(x = "total_bill", y = "tip", hue = "smoker", col = "time", data = df)
```

Bir önceki grafiğe ek olarak 4. bir boyutu ekledik. Ve bunu da ızgara (grid) yöntemi ile yaptık. Izgara yöntemini, farklı kırılımlar ekleyip farklı bakış açsıcı kazanma ve yrum yapaibilme için sıkça kullanmalısın. Çokça fikir verebilir. (R'daki Pair Plot bunun benzeri)


Bir önceki grafikte, total hesabı göz önüne alıp sigara içme-içmeme durumuna göre kişilerin verdiği bahşiş ile total hesap oranına bakmıştık. Burda da aynı grafiğe "time" kırılımını/boyutunu ekledik. Bu Değişken/Özellikte, önceki grafiğin günün hangi yemek saatinde nasıl değiştiğini gözlemliyoruz.

Öğle yemeğine bakıldığında, sigara içen ile içmeyenin verdiği bahşiş ile hesap oranı aynı gibiyken, akşam yemeğinde fark artmaya başlamış.

Bazen de şu yorumu yapabilmekte önemlidir: Evet grafiklere bakıldığında bir fark var ama bu dikkate alınacak kadar değerli mi ?

```python
sns.lmplot(x = "total_bill", y = "tip", hue = "smoker", col = "time", row = "sex",data = df)
```
