# Fundamentals of Matplotlib
Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
```

```python
ageList = [10,20,30,30,30,40,50,60,70,75]
```

```python
weightList=[45,75,80,77,87,75,100,80,75,81]
```

```python
numpAgeList = np.array(ageList)
numWeightList = np.array(weightList)
```

```python
plt.plot(numpAgeList,numWeightList,"b")  #rgb....
```

```python
plt.plot(numWeightList,numpAgeList,"b")
```

```python
plt.plot(numpAgeList,numWeightList,"g")
plt.xlabel("Weight")
plt.ylabel("Age")
plt.title("Change graph of weight by age")
#plt.show
```

### Customizing Graphs

```python
numpArray1= np.linspace(1,10,19)
numpArray1
```

```python
numpArray2 = numpArray1**3
numpArray2
```

```python
plt.plot(numpArray1,numpArray2,"g--")
```

```python
plt.plot(numpArray1,numpArray2,"g*-")
```

```python
plt.plot(numpArray1,numpArray2,"g*--")
```

### Side by Side Graphs

```python
plt.subplot(1,2,1)    ##row,column, rank(first graph)
plt.plot(numpArray1,numpArray2,"r*-")
plt.subplot(1,2,2)    ##second graph
plt.plot(numpArray2,numpArray1,"g--")
```

```python
plt.subplot(2,1,1)    ##row,column, rank(first graph)
plt.plot(numpArray1,numpArray2,"r*-")
plt.subplot(2,1,2)    ##second graph
plt.plot(numpArray2,numpArray1,"g--")
```

### More Customizing
If you want more than the above graphs, you can use figures. Figures like an empty canvas. You can use to plot all the surface.

```python
myFigure = plt.figure()  
# t axes start point, x axes star point,width-lenght
figureAxes = myFigure.add_axes([0.2,0.2,0.9,0.9])
figureAxes.plot(numpArray1,numpArray2,"g")
figureAxes.set_xlabel("X Ekseni Veri İsmi")
figureAxes.set_ylabel("Y Ekseni Veri İsmi")
figureAxes.set_title("Grafik Başlığı")
```

### Advenced Figure => Nested Graphs

```python
myfigure2 = plt.figure()

axes1 = myfigure2.add_axes([0.1,0.1,0.7,0.7])
axes2 = myfigure2.add_axes([0.3,0.3,0.3,0.3])


axes1.plot(numpArray1,numpArray2,"g-")
axes1.set_xlabel("X Axes")
axes1.set_ylabel("Y Axes")
axes1.set_title("Inner Graph Title")

axes2.plot(numpArray1,numpArray2,"b-")
axes2.set_xlabel("X Axes")
axes2.set_ylabel("Y Axes")
axes2.set_title("Inner Graph Title")
```

```python
myfigure2 = plt.figure()

axes1 = myfigure2.add_axes([0.5,0.5,0.7,0.7])
axes2 = myfigure2.add_axes([0.3,0.3,0.3,0.3])


axes1.plot(numpArray1,numpArray2,"g-")
axes1.set_xlabel("X Axes")
axes1.set_ylabel("Y Axes")
axes1.set_title("Inner Graph Title")

axes2.plot(numpArray1,numpArray2,"b-")
axes2.set_xlabel("X Axes")
axes2.set_ylabel("Y Axes")
axes2.set_title("Inner Graph Title")
```

```python
myfigure2 = plt.figure()

axes1 = myfigure2.add_axes([0.1,0.1,0.7,0.7])
axes2 = myfigure2.add_axes([0.4,0.4,0.3,0.3])


axes1.plot(numpArray1,numpArray2,"g-")
axes1.set_xlabel("X Axes")
axes1.set_ylabel("Y Axes")
axes1.set_title("Inner Graph Title")

axes2.plot(numpArray1,numpArray2,"b-")
axes2.set_xlabel("X Axes")
axes2.set_ylabel("Y Axes")
axes2.set_title("Inner Graph Title")
```

=> Also we can do nested graphs with subplot. For this, we should know what subplots returns.

```python
plt.subplots()  ## returns tuple => (figure,axes)
```

```python
(myFigure,myAxes)=plt.subplots()
myAxes.plot(numpArray1,numpArray2,"b")
```

```python
#myFigure    as you seen
#myAxes
```

```python
(myFigure,myAxes)=plt.subplots(nrows=1,ncols=2)
myAxes.plot(numpArray1,numpArray2,"b")
```

As you seen above, if we generate more than one graph, our axes will be np.array. Therefore we should use for loops to plot our graphs.

```python
type(myAxes)
```

```python
(myFigure,myAxes)=plt.subplots(nrows=1,ncols=2)
for axis in myAxes:
    axis.plot(numpArray1,numpArray2,"g")
    axes2.set_xlabel("X Axes")
    axes2.set_ylabel("Y Axes")
plt.tight_layout()   # expand space
```
