# Visual Improving/Enhancements 2

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
numpArray1 = np.linspace(1,10,19)
numpArray1
```

```python
numpArray2 = numpArray1**2
numpArray2
```

```python
(myFigure,myAxes) = plt.subplots()
myAxes.plot(numpArray1,numpArray2,color="#4A15AB",alpha=0.6) #alpha = transparent
myAxes.plot(numpArray2,numpArray1,color="#CFF568")
```

```python
(newFigure,newAxes)= plt.subplots()
newAxes.plot(numpArray1,numpArray1+10,"b",linewidth=5.0)
newAxes.plot(numpArray1,numpArray1+11,"g",linewidth=2.0)
newAxes.plot(numpArray1,numpArray1+12,"r",linestyle="--")
newAxes.plot(numpArray1,numpArray1+13,"y",linestyle=":")
newAxes.plot(numpArray1,numpArray1+14,"p",linestyle="-.")
newAxes.plot(numpArray1,numpArray1+15,"p",marker="+")
## marker, where the x and y axis intersect
newAxes.plot(numpArray1,numpArray1+16,"p-",marker="o",markerfacecolor="r",markersize=8)
```

## Different Type of Graphs

### Scatter plot

```python
plt.scatter(numpArray1,numpArray2)
```

```python
plt.plot(numpArray1,numpArray2,"o")  #as you seen, you can make an same type with scatter
```
