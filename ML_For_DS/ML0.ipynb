## Model Doğrulama Yöntemleri

Modellerin ürettiği sonuçları kullanarak bu sonuçların doğruluğunun değerlendirilmesi/test edilmesidir. Bu yöntemleri kullanma sebebimiz: model başarı sonuçlarının daha doğru değerlendirilmesidir.

### Holdout Yöntemi (Sınama Yöntemi)


En bilindik yöntemdir. Elimizdeki veri setini Eğitim (Train) seti ve Test veri seti olarak ikiye böler, eğitim ile eğitir test ile de test ederiz. Ayırma yöntemlerinde 3/2 eğitim, 3/1 test gibi gibi ayırma yöntemleri var. Bu yöntem veri setindeki gözlemlerin az olduğu durumlarda etkili olamaması gibi bir eksisi vardır. Büyük veri setlerinde ise çok etkilidir. <br>**bold text**

### K Fold Cross Validation (K-Katlı Çapraz Doğrulama)

image.png

Bu yöntemde de eğitim veri seti Train ve Test olmak üzere ikiye ayrılır. Sınama(Holdout) Yönteminden farkı: Eğitim veri seti k adet parçaya bölünür, bir parça seçilip içinden ayrılır ve geri kalan ile eğitim yapılır. Eğitim yapıldıktan sonra kenara ayrılan parça ile model test edilir. Daha sonra bir başka parça seçilir ve kenara ayrılır ve böylece iterayona devam eder. K kere model eğitilirken, k-1 adet parçayla eğitilir ve K kere kendi içinde test edilir. İşte yukarıdaki örnekte bu model kendi içinde 5 kere eğitilir ve bu 5 eğitimin kendi içindeki test sonuçlarının ortalaması bize Validasyon (Doğruluk) Hatasını verir. Önemli olan şey en başta ayırdığımız Test Setini K-Fold'a dahil etmedik, dikkat et. Test Hatası ile Validasyon Hatası aynı şey değildir! Eğitim Setiyle eğitip Test Setiyle test ettiğimde elde ettiğim şey <u>Test Hatası</u> fakat k-fold'daki gibi Eğitim Seti içerisinde parçalara ayırıp Eğitim Setini kendi içerisinde değerlendirdiğimde aldığım şey <u>Validasyon Hatasıdır.</u> <br>
K-Fold'un artısı; Eğitim Setinde elde ettiğimiz Hata Oranı'nı, Test Seti ile Test edip daha doğru bir sonuç bulmamız. <br>
Şunu tekrar etmekte fayda var; Validasyon işlemi Eğitim Veri Seti üstünde yapılır. Hatta siz tüm Veri Setini, Eğitim veri seti olarak kabul edebilir ve çıkan sonucada Validasyon Hatası olarak ele alabilirsiniz. Önemli olan eğitim setinin kendi içerisinde testi. Fakat tüm seti Eğitim Seti yaptığımızda aldığımız sonuca Test Sonucu diyemeyiz, bu aşırı yanlış olur. Test seti öğrenmede kullanılmamlıdır. Yoksa overfittinge sebep olur. <br>
Test Setini ayırıp kullanmamızın sebebide, modelde "daha az yanlık" olur. Buda overfittingin önüne geçmek için harika bir yoldur. <br>
İntte şöyle bir yazı buldum, işe yarayabilir: \
*Validation bölümü train veri seti içinden seçilir. Train veri seti üzerinde doğru model seçimi yapılarak algoritma belirlenir. Validation bölümünde ise uygulanan model iyileştirilmeye çalışılır. Bunun için hiper parametrik (hyperparameter tuning) uygulamalar denenerek en optimum katsayılar/ağırlıklar bulunmaya çalışılır. Özellikle çok büyük veri setleri üzerinde sürekli Train datası üzerinden çalışılamayacağı için, küçük bir bölüm alınarak validation olarak tanımlanır.*

## Parametre, Hiperparametre, Prametre Tunning Model Optimizasyonu
Model tahmin performansını iyileştirmeyi işleyeceğiz. <br>
Başlamadan önce şunu bilmekte fayda var; Model Parametreleri, modelin ürettiği parametrelerdir. Model Hiperparametreleri ise modelin üretmediği, dışsal parametrelerdir. İşte bu bölümde yukarıdaki parametrelerin ilşkisini inceleyeceğiz. Tunning, ayarlamak demek, parametre ayarlama-model ayarlama.

### Gradient Descent (Gradyan Azaltılması/İnişi):

Modelimizde bir hata oranı bulduk, MSE, RMSE gibi gibi yollarla. Fakat bu hatalara bakmak içinde bulmadık. Bu hataları azaltıp modelimizi daha optimize hale getirmeye çalışacağız. Hataları azaltma/ modeli optimize etme çabalarımızda gradient descent kullanılıyor. <br>
Temelde her şey matematiktir. Burada da kullanacğımız işin temelinde matematik var ve en öenmli konusu, Türev. Konuya şöyle girelim: Gradient Descent iteratif bir fonksiyondur, yani adım adım ilerler. Belirlediğimiz optimum noktaya (daha sonra detaylı anlatıalcak) gelince de durur ve o değeri seçer, bu noktları kıyaslarken türevini alır ve birbirleri ile kıyaslar. Hatırla, türev bize o noktadaki eğimi verirdi. Gradient Descent'tede noktalrın türevlerine karşılık gelen değerleri kıyaslıyoruz. Seninde tahmin edeceğin üzere eğimin sıfır olduğu yer bizim için en optimum nokta, çünkü fonksiyonun aşşağıdan yuakrıya çıktığı yer oluyor. EN OPTİMUM WEIGHT DEĞRLERİ BULUNUR. Böylece hata oranı azaltılır.

image.png

Bazen çıkan fonksiyonda birden fazla min noktası, yani türevin 0'a eşit olduğu noktalar olabilir. Bunu yerel max - global max değerleri olara kliseden hatırlarsın. Modelimiz yerel minlerden birine takılıp kalabilir. Bunun önüne geçmekiçin çeşitli grad desc türleri üretilmiştir. Bunlar:

#### Batch Grad Desc

Tüm veri setinin 1 epochta verildiği grad desc türüdür.  1 iterasyon sürer tüm set 1 epochta verildiği için fakat ağır ve yavaş işler. Zaman alır fakat global mine yaklaşma olasılığı daha fazladır.

#### Mini Batch Grad Desc

Bir iterasyonda tüm veri setini vermek yerine parça parça verir, bir kaç iterasyonda sonuca ulaşır. Daha hızlıdır, doğruluk olarak batch grad kadar değildir ama.

#### Stochastic Grad Desc

Rastgele ilerleyerek optimum noktaya ulaşmaya çalışır.

---

Hız Karşılaştırılmaları: <br>
Mini Batch > Stoc > Grad
<br> <br>
Optimuma Ulaşma: <br>
Grad > Mini Batch > Stoc

### Hiperparametreler

Yukarıda modelin optimizasyonu ile ilgili olan Gradient Descent'i inceledik. Şimdi de detaylara girme vakti. Hiperparametreler, modelimizin ürettiği parametreler değildir. Bu parametreler ile modele etki ederiz. Bu hiperparametreler: epoch, iterayon, learning rate. Bunlarla oynayarakta (yani grad descin yanı sıra) model optimizasyonunu gerçekleştiriyoruz.

#### Epoch ve Iterasyon:

* Iterasyon adım anlamına gelir. Epoch ise tüm veri setinin input olarak modele verilmesi. Batch size ise 1 iterasyona verilen veri miktarı. Şöyle düşünelim: bir kitap var ve 1000 sf. Bu kitabı bir oturuşta bitirirsek, 1 iterasyonda 1 epoch bitirmiş oluruz ve bacth size'ımız 1000. Kitabı 100'er sayfalar halinde okuyacaksak o zaman iterasyonumuz 10 ve batch size'ımız 100 olur. Not:Bir model eğitilirken birden fazla epoch döner. O yüzden epoch değeride önemlidir. Bazı yerlerde bize eğitim süresi hakkında değer olur, eğitim 1000 epoch sürecek gibi gibi.

#### Learning Rate:

* Gradyan inişinde kullanılan bir hiper parametredir. İnerken bir noktadan diğer noktaya atılacak olan adım boyutunu belirler. Bu yüzden büyük learning ratelerde optimum noktayı kaçırabiliriz, aşırı küçük değerlerde de optimizasyon yavaş sürebilir.
