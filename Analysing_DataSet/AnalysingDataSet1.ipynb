### First: Fill with 0

```python
df2 = df.copy()
df2["orbital_period"].fillna(0,inplace=True)
df2.isnull().sum()
```

### Second: Fill with Mean

```python
df2["mass"].fillna(df2["mass"].mean(),inplace=True)
df2.isnull().sum()
```

### Third: Fill with Median

```python
df2.distance.fillna(df2["distance"].median(),inplace=True)
df2.isnull().sum()
```
